```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
df.plot.scatter(x='A', y='B', color='DarkBlue', label='Class 1')
df.plot.scatter(x='A', y='C', color='LightGreen', label='Class 2')
df.plot.scatter(x='A', y='D', color='DarkBlue', label='Class 1')
df.plot.scatter(x='A', y='B', color='DarkBlue', label='Class 1')
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
df.plot.scatter(x='A', y='B', color='DarkBlue', label='Class 1')
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
df.plot.scatter(x='A', y='B', color='DarkBlue', label='Class 1')
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('20130101',periods=10),columns=list('ABCD'))
df.plot()
plt.show()
```

```python
#
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'],index=pd.date_range('20130101',periods=10))
df.plot.bar()
plt.show()
```
