```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 设置字体
plt.rcParams['font.sans-serif'] =' SimHei'
# 设置中文显示
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("./data/people.csv", encoding="utf-8")
name = data.columns
values = data.values

plt.figure(figsize=(12, 8))
plt.scatter(values[:, 0], values[:, 1], c="g", marker="o", alpha=0.5)
plt.grid()
plt.xlabel("年份")  # 添加横轴标签
plt.ylabel("年末总人口（万人）")  # 添加y轴名称
plt.xticks(range(0, 20), values[range(0, 20), 0], rotation=45)
plt.title("2000-2019年年末总人口散点图")  # 添加图表标题
plt.show()
```
