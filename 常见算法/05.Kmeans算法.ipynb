**1.基础代码演示**

```python
#创建数据集
import numpy as np
data = np.array([[3, 5], [5, 4], [5, 9],  [7, 6],[4, 2], [3, 2], [4, 1], [3, 6], [5, 8],[4, 7], [3, 9], [7, 5], [6, 8], [6, 6], [7, 7]])
# data = np.random.randint(1,100,[20,2]) #(1,100）以内的[行,列]随机整数

data
```

```python
#图像展示
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(data[:, 0], data[:, 1], c="b", marker='o', label='data')  
plt.legend()  
plt.show()
```

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=2)
kms.fit(data)
```

```python
label = kms.labels_
print(label)
```

```python
plt.scatter(data[label == 0][:, 0], data[label == 0][:, 1], c="b", marker='o', label='A') 
plt.scatter(data[label == 1][:, 0], data[label == 1][:, 1], c="r", marker='*', label='B') 
plt.legend()  # 设置图例
```

```python
kms_3 = KMeans(n_clusters=3)
kms_3.fit(data)
label_3 = kms_3.labels_
print(label_3)
```

```python
plt.scatter(data[label_3 == 0][:, 0], data[label_3 == 0][:, 1], c="b", marker='o', label='A') 
plt.scatter(data[label_3 == 1][:, 0], data[label_3 == 1][:, 1], c="r", marker='*', label='B') 
plt.scatter(data[label_3 == 2][:, 0], data[label_3 == 2][:, 1], c="g", marker='+', label='C')  
plt.legend() 
```

**2.小案例 - 考试成绩划分**

**3.用户分群**

```python
# 1.读取数据
import pandas as pd 
# data = pd.read_excel('D:\学习\python\jupyter\商分-机器学习课程相关代码\dataset\用户数据.xlsx')
data = pd.read_csv('./dataset/用户数据 - 副本.csv')
data.head()


# 2.划分X和y

import matplotlib.pyplot as plt
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c="pink", marker='+') 
plt.xlabel('age')  # 添加x轴名称
plt.ylabel('gmv')  # 添加y轴名称
plt.show()


# 3.划分训练集和测试集  --无监督不用



# 4.模型训练
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=3, random_state=123)
kms.fit(data)
label = kms.labels_
label = kms.fit_predict(data)
print(label)

```

```python
plt.scatter(data[label == 0].iloc[:, 0], data[label == 0].iloc[:, 1], c="pink", marker='o', label='A')  
plt.scatter(data[label == 1].iloc[:, 0], data[label == 1].iloc[:, 1], c="green", marker='*', label='B')  
plt.scatter(data[label == 2].iloc[:, 0], data[label == 2].iloc[:, 1], c="orange", marker='+', label='C')  
plt.xlabel('age')  # 添加x轴名称
plt.ylabel('gmv')  # 添加y轴名称
plt.legend()  
```
