```python
import numpy as np
from scipy.optimize import minimize

# 定义目标函数
def objective(x):
    return x[0]**2 + x[1]**2

# 初始猜测值
x0 = np.array([1.0, 1.0])

# 最小化目标函数
result = minimize(objective, x0)

print("优化结果:", result.x)
print("目标函数最小值:", result.fun)

# 画出优化过程
x = np.arange(-2, 2, 0.01)
y = x**2
plt.plot(x, y)
plt.plot(result.x[0], result.fun, "r*")
plt.show()

# 定义目标函数
def objective(x):
    return x[0]**2 + x[1]**2
```
